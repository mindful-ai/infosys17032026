```python
class Apparel:

    Counter = 100
    
    def __init__(self, price, item_type):
        Apparel.Counter += 1
        self.__price = price
        self.__item_type = item_type
        if item_type in ["Silk", "Cotton"]:
            if(item_type == "Silk"):
                self.__itemid = "S" + str(Apparel.Counter)
            else:
                self.__itemid = "C" + str(Apparel.Counter)
        else:
            raise Exception("Invalid Apparel Type")

    def set_itemid(self, id):
        self.__itemid = id

    def get_itemid(self):
        return self.__itemid

    def set_price(self, price):
        self.__price = price

    def get_price(self):
        return self.__price

    def set_item_type(self, type):
        self.__item_type = type

    def get_item_type(self):
        return self.__item_type

    def calculate_price(self):
        self.__price = self.__price + (self.__price * 0.05)
        
```

```python
class Cotton(Apparel):

    def __init__(self, price, discount):
        super().__init__(price, "Cotton")
        self.__discount = discount 

    def __str__(self):
        return str(super().get_itemid()) + ' --> '+ str(super().get_item_type()) + ' --> ' + str(super().get_price())

    def get_discount(self):
        return self.__discount

    def calculate_price(self):
        super().calculate_price()
        discountedPrice = super().get_price() - self.__discount
        vatAppliedPrice = discountedPrice + (discountedPrice * 0.05)
        super().set_price(vatAppliedPrice)
        
    
```

```python
class Silk(Apparel):

    def __init__(self, price):
        super().__init__(price, 'Silk')
        if(price > 10000):
            self.__points = 10
        else:
            self.__points = 3
            
    def __str__(self):
        return str(super().get_itemid()) + ' --> '+ str(super().get_item_type()) + ' --> ' + str(super().get_price())

    def get_points(self):
        return self.__points

    def calculate_price(self):
        super().calculate_price()
        vatAppliedPrice = super().get_price() + (super().get_price() * 0.1)
        super().set_price(vatAppliedPrice)

    
```
