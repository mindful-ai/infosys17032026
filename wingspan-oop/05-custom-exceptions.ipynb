```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(5)
```

```python
factorial(-1)
```

### Custom Exception and Usage

```python
class NegativeValueError(Exception):

    def __init__(self, message):
        super().__init__(message)
```

```python
def factorial(n):
    if n <= 0:
        raise NegativeValueError('Negative values cannot be specified for factorial')
    else:
        if n == 1:
            return 1
        else:
            return n * factorial(n - 1)
```

```python
factorial(-1)
```

```python
try:
    factorial(-1)
except Exception as e:
    print(e)
```

### Another Example

```python
class InvalidSalaryException(Exception):

    def __init__(self, salary, message="Salary should be greater than 0"):
        self.salary = salary
        self.message = message
        super().__init__(self.message)

    def __str__(self):
        return f"{self.message} -> Given: {self.salary}"
```

```python
class Employee:

    def __init__(self, name, company, salary):
        if(salary <= 0):
            raise InvalidSalaryException(salary)
        self.salary = salary
        self.name = name
        self.company = company

    def get_details(self):
        return self.name, self.company, self.salary
```

```python
e1 = Employee("Anil", "Wipro", 1000000)
e1.get_details()
```

```python
e1 = Employee("Sunil", "Wipro", -5000)
e1.get_details()
```

```python
class Product:
    counter=100
    def __init__(self,name,description,brand,price):
        Product.counter+=1
        self.id='P'+str(Product.counter)
        self.name=name
        self.description=description
        self.brand=brand
        self.price=price

    # Operator Overloading
    def __add__(self, other):
        """
            a + b
            self -> LHS object -> a
            other -> RHS object -> b
        """
        return (str(self.name) + ' ' + str(self.description))  + ' --  ' + \
               (str(other.name) + ' ' + str(other.description))
```

```python
product1 = Product('IPhone','Smart Phone','Apple',72500.25)
product2 = Product('Air Max','A Sport Shoe','Nike',1280.0)
print(product1 + product2)
```
