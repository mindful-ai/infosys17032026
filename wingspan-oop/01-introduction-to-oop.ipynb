### 1.2 Creating the classes and objects correctly

```python
class Student:

    # Constructor
    def __init__(self):
        pass

    # Destructor
    def __del__(self):
        pass
```

```python
class Student:

    # Constructor
    def __init__(self, name, age, grade):
        # Instance Variables
        self.name = name
        self.age  = age
        self.grade = grade

    # Destructor
    def __del__():
        print("Job Completed!!")

    # Member Function
    def printinfo(self):
        print("SELF  -> ", self)
        print(self.name, self.age, self.grade)
```

##### Inbound and Outbout Invocation

```python
class Mobile:
    def __init__(self):
        print("Inside constructor")
    def purchase (self):
        print("Purchasing a mobile")
mob1=Mobile()
mob1.purchase() #Inbound method invocation, We need not pass the value for self.
Mobile.purchase(mob1) #Outbound method invocation, We have to pass the object as the value of self.
```

### 1.3 Exercise - Solution

```python
class Vehicle:

    # attributes
    def __init__(self, vid, vtype, vcost):
        self.vid = vid
        self.vtype = vtype
        self.vcost = vcost
        
    # methods
    def calculatePremium(self):
        if self.vtype in ['Two Wheeler', 'Four Wheeler']:
            return 0.02 * self.vcost if self.vtype == 'Two Wheeler' else 0.06 * self.vcost

    def printVehicleInfo(self):
        print("VEHICLE ID    : ", self.vid)
        print("VEHICLE TYPE  : ", self.vtype)
        print("VEHICLE COST  : ", self.vcost)
        print("-"*80)
        print("PREMIUM       : ", self.calculatePremium())
```

```python
v1 = Vehicle("KA41Z8896", "Four Wheeler", 1500000)
v2 = Vehicle("KA30H1892", "Two Wheeler", 100000)
```

```python
v1.printVehicleInfo()
```

```python
v2.printVehicleInfo()
```

### 1.4 Encapsulation

```python
class Customer:
    def __init__(self, cust_id, name, age, wallet_balance):
        self.cust_id = cust_id
        self.name = name
        self.age = age
        self.wallet_balance = wallet_balance
    def update_balance(self, amount):
        if amount < 1000 and amount > 0:
            self.__wallet_balance += amount
    def show_balance(self):
        print ("The balance is ",self.__wallet_balance)

```

```python
c1=Customer(100, "Gopal", 24, 1000)
print(c1.wallet_balance)
```

```python
c1.wallet_balance = 10000000
```

```python
print(c1.wallet_balance)
```

#### How to make wallet_balance private in python classes?

```python
class Customer:
    def __init__(self, cust_id, name, age, wallet_balance):
        self.cust_id = cust_id
        self.name = name
        self.age = age
        self.__wallet_balance = wallet_balance ## private variable creation
    def update_balance(self, amount):
        if amount < 1000 and amount > 0:
            self.__wallet_balance += amount
    def show_balance(self):
        print ("The balance is ",self.__wallet_balance)
```

```python
c1=Customer(100, "Gopal", 24, 1000)
print(c1.__wallet_balance)
```

```python
c1.show_balance()
```

```python
c1.__wallet_balance = 10000000
```

```python
c1.show_balance()
```

### 1.5 Static Attributes and Methods

```python
class Student:

    # static variable, class variable
    __nStudents = 0
    
    # Constructor
    def __init__(self, name, age, grade):
        # Instance Variables
        self.name = name
        self.age  = age
        self.grade = grade
        self.subjects = {}
        self.average = 0
        Student.__nStudents += 1

    # static methods
    @staticmethod
    def get_nstudents():
        return Student.__nStudents

    @staticmethod
    def avgFunction(*values):
        return sum(values)/len(values)

    # setters and gettesr
    def set_subjects(self, marks):
        self.subjects = marks

    def calc_average(self):
        self.average = Student.avgFunction(*self.subjects.values())
        

    # Other member Function
    def printinfo(self):
        print("SELF  -> ", self)
        print(self.name, self.age, self.grade)
        print('-'*50)
        print(self.subjects)
        print(self.average)
```

```python
s1 = Student("Raj", 12, "6th")
s2 = Student("Ram", 13, "7th")
```

```python
Student.get_nstudents()
```

```python
s1.set_subjects({'phy':89, 'math':99})
```

```python
s1.printinfo()
```

```python
s1.calc_average()
```

```python
s1.printinfo()
```
