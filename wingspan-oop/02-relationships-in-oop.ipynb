### 2.1 Aggregates -> "has-a" relationship

```python
class Customer:
    def __init__(self, name, age, phone_no, address):
        self.name = name
        self.age = age
        self.phone_no = phone_no
        self.address = address
    def view_details(self):
        print (self.name, self.age, self.phone_no)
        print (self.address.door_no, self.address.street, self.address.pincode)
    def update_details(self, addr):
        self.address = addr
```

```python
class Address:
    def __init__(self, door_no, street, pincode):
        self.door_no = door_no
        self.street = street
        self.pincode = pincode
    def update_address(self, addr):
        self.door_no = addr['door_no']
        self.street = addr['street']
        self.pincode = addr['pincode']
```

```python
addr_01 = Address("#2345", "M G Road, Bangalore", "560001")
```

```python
cust_01 = Customer("Anil Kumar", 45, "+917786534211", addr_01)
```

```python
cust_01.view_details()
```

### 2.2 Association - Class Dependency - Not Strict

```python
class Customer:
    def __init__(self, name, age, phone_no):
        self.name = name
        self.age = age
        self.phone_no = phone_no
    def purchase(self, payment):
        if payment.type == "card":
            print ("Paying by card")
        elif payment.type == "e-wallet":
            print ("Paying by wallet")
        else:
            print ("Paying by cash")

```

```python
class Payment:
    def __init__(self, type):
        self.type = type
```

```python
payment1 = Payment("card")
c = Customer("Jack",23,1234)
c.purchase(payment1)
```

### 2.3 Creating an object inside an object

```python
#Object creation
class Customer:
    def __init__(self, name,cust_type,bill):
        self.name = name
        self.bill = bill
        self.cust_type=cust_type
    def calculate_bill(self):
        tax1 = Tax(self.cust_type)
        final_bill = self.bill - tax1.tax_details(self.cust_type)
        return final_bill

```

```python
class Tax:
    def __init__(self,cust_type):
        self.cust_type=cust_type
    def tax_details(self,cust_type):
        if(cust_type=="Student"):
            return 5
        else:
            return 10
```

```python
cust1=Customer("Maddy","Senior Citizen",100)
print(cust1.calculate_bill())
```

### 2.4 Using static attributes and methods

```python
#Usage of static method
class CustomerCare:
    __helpline=111000
    @staticmethod
    def get_helpline():
        return CustomerCare.__helpline

```

```python
class Customer:
    def call_support(self):
        print("Calling ",CustomerCare.get_helpline())
```

```python
Customer().call_support()
```
